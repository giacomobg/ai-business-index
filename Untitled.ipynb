```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import os
os.listdir()
```

```python
df = pd.read_csv('data-1.csv')
df.head()
```

```python
df.groupby('Region').mean()
```

```python
fig, ax = plt.subplots(4,4,figsize=(10,10))
variables = ['Deals', 'Fillings', 'Jobs', 'Patents']
for count1, var1 in enumerate(variables):
    for count2, var2 in enumerate(variables):
        ax[count1,count2].plot(df[var1], df[var2], '.')
        ax[count1,count2].set_title(var2 + ' v '+ var1)
        ax[count1,count2].set_xlim(1,3)
        ax[count1,count2].set_ylim(1,3)
plt.show()
```
